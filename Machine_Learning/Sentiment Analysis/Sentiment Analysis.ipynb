```python
%matplotlib inline
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
from IPython.display import display

yelp_train_raw = pd.read_csv('./Datasets/yelp-train.txt', sep='\t', names=["Review","Rating"])
yelp_valid_raw = pd.read_csv('./Datasets/yelp-valid.txt', sep='\t', names=["Review","Rating"])
yelp_test_raw  = pd.read_csv('./Datasets/yelp-test.txt', sep='\t', names=["Review","Rating"])

movie_train_raw = pd.read_csv('./Datasets/IMDB-train.txt', sep='\t', names=["Review","Sentiment"])
movie_valid_raw = pd.read_csv('./Datasets/IMDB-valid.txt', sep='\t', names=["Review","Sentiment"])
movie_test_raw  = pd.read_csv('./Datasets/IMDB-test.txt', sep='\t', names=["Review","Sentiment"])

# features
yelp_X_train = yelp_train_raw['Review'].values
yelp_X_valid = yelp_valid_raw['Review'].values
yelp_X_test  = yelp_test_raw['Review'].values

movie_X_train = movie_train_raw['Review'].values
movie_X_valid = movie_valid_raw['Review'].values
movie_X_test  = movie_test_raw['Review'].values

# target labels
yelp_y_train = yelp_train_raw['Rating'].values
yelp_y_valid = yelp_valid_raw['Rating'].values
yelp_y_test  = yelp_test_raw['Rating'].values

movie_y_train = movie_train_raw['Sentiment'].values
movie_y_valid = movie_valid_raw['Sentiment'].values
movie_y_test  = movie_test_raw['Sentiment'].values
```

### 1.  Raw Text Vectorization
Convert the varying length review into a fixed length vector through:
- Binary bag-of-words
- Frequency bag-of-words

#### 1.1 Input Preprocessing
Form a set of words after removing ponctuation and casting to lower-case:

```python
import string
from collections import Counter

no_punct_trans = str.maketrans({key: None for key in string.punctuation})

# remove punctuation from string and cast to lower case
get_clean_words = lambda text: text.translate(no_punct_trans).lower()

# form a single string containing all the concatenated reviews
concat_column = lambda col: col.str.cat(sep=' ')

yelp_clean_words  = get_clean_words(concat_column(yelp_train_raw['Review']))
movie_clean_words = get_clean_words(concat_column(movie_train_raw['Review']))

yelp_word_freq  = Counter(list(yelp_clean_words.split()))
movie_word_freq = Counter(list(movie_clean_words.split()))

print('\nUnique words:\t Yelp: {}\t IMDB: {}'.format(len(yelp_word_freq), len(movie_word_freq)))
```

Generate a sorted list of words in descending order of frequency:

```python
import operator

def get_n_frequent_words(word_freq, n):
    sorted_words = sorted(word_freq.items(),
                          key=operator.itemgetter(1),
                          reverse=True)
    return sorted_words[:n]

M = 10000
    
yelp_most_freq  = get_n_frequent_words(yelp_word_freq, M)
movie_most_freq = get_n_frequent_words(movie_word_freq, M)
```

Store the vocabulary as required by the assignment

```python
def write_vocab(word_freq, filename):
    with open(filename, 'w', encoding='utf-8') as f:
        for idx, (word, freq) in enumerate(word_freq):
            f.write('{}\t{}\t{}\n'.format(word, idx, freq))
    
write_vocab(yelp_most_freq, './Output/yelp-vocab.txt')
write_vocab(movie_most_freq, './Output/IMDB-vocab.txt')
```

Get all the non-zero features for each of the training sets:

```python
%%time
# fill a vector feature with an input text 
def get_bin_features(text, key_words):
    bins = []
    cnt = Counter(np.array(get_clean_words(text).split()))
    for idx, key in enumerate(key_words):
        if key in cnt:
            bins.append(idx)
    
    return bins

def fill_bin_features(matrix, data, feats):
    for idx, review in enumerate(data):
        matrix[idx] = get_bin_features(review, feats)

        
yelp_nonzero_train = [0] * len(yelp_train_raw)
yelp_nonzero_valid = [0] * len(yelp_valid_raw)
yelp_nonzero_test  = [0] * len(yelp_test_raw)

movie_nonzero_train = [0] * len(movie_train_raw)
movie_nonzero_valid = [0] * len(movie_valid_raw)
movie_nonzero_test  = [0] * len(movie_test_raw)

yelp_feats  = [item[0] for item in yelp_most_freq]
movie_feats = [item[0] for item in movie_most_freq]

print('\nBin-Bag Vectorizing yelp ...')
fill_bin_features(yelp_nonzero_train, yelp_X_train, yelp_feats)
fill_bin_features(yelp_nonzero_valid, yelp_X_valid, yelp_feats)
fill_bin_features(yelp_nonzero_test, yelp_X_test, yelp_feats)

print('\nBin-Bag Vectorizing IMDB ...\n')
fill_bin_features(movie_nonzero_train, movie_X_train, movie_feats)
fill_bin_features(movie_nonzero_valid, movie_X_valid, movie_feats)
fill_bin_features(movie_nonzero_test, movie_X_test, movie_feats)
```

Store the train/valid/test files as required by the assignment:

```python
def write_dataset(filename, data):
    with open(filename, 'w', encoding='utf-8') as f:
        for (features, label) in data:
            for feat in features:
                f.write('{} '.format(feat))
            f.write('\t{}\n'.format(label))
            
yelp_train = np.vstack((yelp_nonzero_train, yelp_y_train)).T
yelp_valid = np.vstack((yelp_nonzero_valid, yelp_y_valid)).T
yelp_test  = np.vstack((yelp_nonzero_test, yelp_y_test)).T

movie_train = np.vstack((movie_nonzero_train, movie_y_train)).T
movie_valid = np.vstack((movie_nonzero_valid, movie_y_valid)).T
movie_test  = np.vstack((movie_nonzero_test, movie_y_test)).T

write_dataset('./Output/yelp-train.txt', yelp_train)
write_dataset('./Output/yelp-valid.txt', yelp_valid)
write_dataset('./Output/yelp-test.txt', yelp_test)

write_dataset('./Output/IMBD-train.txt', movie_train)
write_dataset('./Output/IMBD-valid.txt', movie_valid)
write_dataset('./Output/IMBD-test.txt', movie_test)
```

Generate a sparse matrix representation of the datasets to facilitate its manipulation:

#### 1.2 Binary Bag of words:

```python
def to_sparse_bin_bag(vectorized_feats, shape):
    full_matrix = np.zeros(shape)
    for idx, row in enumerate(vectorized_feats):
        sample = np.zeros((1, shape[1]))
        sample[0, row] = 1
        full_matrix[idx] = sample
    
    return csr_matrix(full_matrix)


yelp_X_bin_train = to_sparse_bin_bag(yelp_nonzero_train, (len(yelp_nonzero_train), M))
yelp_X_bin_valid = to_sparse_bin_bag(yelp_nonzero_valid, (len(yelp_nonzero_valid), M))
yelp_X_bin_test  = to_sparse_bin_bag(yelp_nonzero_test, (len(yelp_nonzero_test), M))

movie_X_bin_train = to_sparse_bin_bag(movie_nonzero_train, (len(movie_nonzero_train), M))
movie_X_bin_valid = to_sparse_bin_bag(movie_nonzero_valid, (len(movie_nonzero_valid), M))
movie_X_bin_test  = to_sparse_bin_bag(movie_nonzero_test, (len(movie_nonzero_test), M))
```

#### 1.3 Frequency Bag of words:

```python
%%time
# fill a vector feature with an input text 
def get_freq_features(text, key_words):
    freqs = {}
    cnt = Counter(np.array(get_clean_words(text).split()))
    for idx, key in enumerate(key_words):
        if key in cnt:
            freqs[idx] = cnt[key]
    
    return freqs

def fill_freq_features(matrix, data, feats):
    for idx, review in enumerate(data):
        matrix[idx] = get_freq_features(review, feats)

yelp_nonzero_train = [0] * len(yelp_train_raw)
yelp_nonzero_valid = [0] * len(yelp_valid_raw)
yelp_nonzero_test  = [0] * len(yelp_test_raw)

movie_nonzero_train = [0] * len(movie_train_raw)
movie_nonzero_valid = [0] * len(movie_valid_raw)
movie_nonzero_test  = [0] * len(movie_test_raw)

print('\nFreq-Bag Vectorizing yelp ...')
fill_freq_features(yelp_nonzero_train, yelp_X_train, yelp_feats)
fill_freq_features(yelp_nonzero_valid, yelp_X_valid, yelp_feats)
fill_freq_features(yelp_nonzero_test, yelp_X_test, yelp_feats)

print('\nFreq-Bag Vectorizing IMDB ...\n')
fill_freq_features(movie_nonzero_train, movie_X_train, movie_feats)
fill_freq_features(movie_nonzero_valid, movie_X_valid, movie_feats)
fill_freq_features(movie_nonzero_test, movie_X_test, movie_feats)
```

create a sparse matrix representation of the features:

```python
def to_sparse_freq_bag(vectorized_feats, shape):
    full_matrix = np.zeros(shape)
    for idx, row in enumerate(vectorized_feats):
        sample = np.zeros((1, shape[1]))
        for feat, freq in row.items():
            sample[0, feat] = freq
            
        full_matrix[idx] = sample
    
    return csr_matrix(full_matrix)

yelp_X_freq_train = to_sparse_freq_bag(yelp_nonzero_train, (len(yelp_nonzero_train), M))
yelp_X_freq_valid = to_sparse_freq_bag(yelp_nonzero_valid, (len(yelp_nonzero_valid), M))
yelp_X_freq_test  = to_sparse_freq_bag(yelp_nonzero_test, (len(yelp_nonzero_test), M))

movie_X_freq_train = to_sparse_freq_bag(movie_nonzero_train, (len(movie_nonzero_train), M))
movie_X_freq_valid = to_sparse_freq_bag(movie_nonzero_valid, (len(movie_nonzero_valid), M))
movie_X_freq_test  = to_sparse_freq_bag(movie_nonzero_test, (len(movie_nonzero_test), M))
```

### 2.  Yelp & Binary Bag-of-Words
For each of the following classifiers, we will report their validation performance (F1 score).

```python
from sklearn.metrics import f1_score
from sklearn.dummy import DummyClassifier

from itertools import count
iid = count()

predict = lambda clf, Xs: tuple(map(lambda x: clf.predict(x), Xs))

def map_hyperparam(clf, hyperparam_name, hyperparams, Xs, ys):
    train_scores, valid_scores = [], []
    X_train, X_valid = Xs
    y_train, y_valid = ys
    
    for hp in hyperparams:
        clf.set_params(**{hyperparam_name: hp})
        clf.fit(X_train.tocsc(), y_train)

        y_train_pred, y_valid_pred = predict(clf, (X_train, X_valid))

        train_scores.append(f1_score(y_train, y_train_pred, average='micro'))
        valid_scores.append(f1_score(y_valid, y_valid_pred, average='micro'))
    
    return train_scores, valid_scores

def print_performance(y_trues, y_preds):
    print('\nTrain F1 score:\t{:.3f}'.format(f1_score(y_trues[0], y_preds[0], average='micro')))
    print('Valid F1 score:\t{:.3f}'.format(f1_score(y_trues[1], y_preds[1], average='micro')))
    print('Test F1 score:\t{:.3f}'.format(f1_score(y_trues[2], y_preds[2], average='micro')))
    
def plot_performance(xs, train_scores, valid_scores, xlabel):
    fig, ax = plt.subplots(figsize=(10,10))

    ax.plot(xs,train_scores, label='Train score')
    ax.plot(xs,valid_scores, label='Valid score')

    plt.title('Model Performance', fontsize=25)
    plt.xlabel(xlabel, fontsize=20, labelpad=10)
    plt.ylabel('F1 Score', fontsize=20, labelpad=10)

    ax.tick_params(labelsize=15)  

    plt.legend(prop={'size': 20})
    
    # Don't allow the axis to be on top of your data
    ax.set_axisbelow(True)

    # Turn on the minor TICKS, which are required for the minor GRID
    ax.minorticks_on()

    # Customize the major grid
    ax.grid(which='major', linestyle='-', linewidth='0.5', color='black')
    # Customize the minor grid
    ax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
    
    plt.show()

    fig.savefig('./Output/{}.png'.format(next(iid)))
```

Only increasing the maximum depth of the tree will simply result in higher overfitting.  
We must modify the max depth in conjunction with other hyperparameters in order to improve the model's performance.

```python
clf = DT(min_samples_leaf=5)
depths = np.linspace(3,10,8)

Xs = (yelp_X_bin_train, yelp_X_bin_valid)
ys = (yelp_y_train, yelp_y_valid)

train_scores, valid_scores = map_hyperparam(clf, 'max_depth', depths, Xs, ys)

plot_performance(depths, train_scores, valid_scores, 'Max Depth')
```

Next we will analyze the effect of the parameter controlling the minimum samples per leaf.  
Interestingly, we considerably reduce overfitting by not allowing a leaf to be formed unless it has a minimum of 145 samples or more.

```python
clf = DT()
min_samples = np.linspace(10,300,20, dtype=np.int32)

Xs = (yelp_X_bin_train, yelp_X_bin_valid)
ys = (yelp_y_train, yelp_y_valid)

train_scores, valid_scores = map_hyperparam(clf, 'min_samples_leaf', min_samples, Xs, ys)
    
plot_performance(min_samples, train_scores, valid_scores, 'Minimum Samples per Leaf')
```

We can also perform **feature selection**.  
We will attempt to remove useless features via tree-based feature selection:

```python
from sklearn.ensemble import ExtraTreesClassifier as ExtraDT
from sklearn.feature_selection import SelectFromModel

print('\nNum. of feats. in training dataset: {}'.format(yelp_X_bin_train.shape[1]))

clfs = ExtraDT()
clfs = clfs.fit(yelp_X_bin_train.tocsc(), yelp_y_train)

model = SelectFromModel(clfs, prefit=True)

X_train = model.transform(yelp_X_bin_train.tocsc())
X_valid = model.transform(yelp_X_bin_valid.tocsc())
X_test  = model.transform(yelp_X_bin_test.tocsc())

print('Num. of feats. in training dataset after Feat. Selection: {}'.format(X_train.shape[1]))
```

Notice we obtain the same clasification performance using the feature selected dataset as with the original one. But the time it takes to train and classify is reduced:

```python
%%time
clf = DT(max_depth=3)
clf.fit(X_train.tocsc(), yelp_y_train)

Xs = (X_train, X_valid, X_test)
y_trues = (yelp_y_train, yelp_y_valid, yelp_y_test)
y_preds = predict(clf, Xs)

print_performance(y_trues, y_preds)
```

We will now test all possible permutations of maximum depth and minimum number of samples in a leaf:

```python
clf = DT()
min_samples = np.linspace(10,300,20, dtype=np.int32)
depths = np.linspace(5,10,4, dtype=np.int32)

Xs = (X_train, X_valid)
ys = (yelp_y_train, yelp_y_valid)

for depth in depths:
    clf.set_params(max_depth=depth)
    train_scores, valid_scores = map_hyperparam(clf, 'min_samples_leaf', min_samples, Xs, ys)
    
    plot_performance(min_samples, train_scores, valid_scores, 
                     'Min Samples per Leaf with Max Depth: {}'.format(depth))
```

2 potential candidates are identified:
 - Max depth: 6, Min. 190 samples in leaf: 
 - Max depth: 7, Min. 100 samples in leaf: 
 
We choose *{max_depth:6, min_samples_leaf:190}*, but present performances for both hyperparameter choices below:

```python
print('\nMax Depth: 6\tMin. Samples per Leaf: 190')
clf = DT(max_depth=6, min_samples_leaf=190)
clf.fit(X_train.tocsc(), yelp_y_train)

Xs = (X_train, X_valid, X_test)
y_trues = (yelp_y_train, yelp_y_valid, yelp_y_test)
y_preds = predict(clf, Xs)

print_performance(y_trues, y_preds)


print('\nMax Depth: 7\tMin. Samples per Leaf: 100')
clf = DT(max_depth=7, min_samples_leaf=100)
clf.fit(X_train.tocsc(), yelp_y_train)

Xs = (X_train, X_valid, X_test)
y_trues = (yelp_y_train, yelp_y_valid, yelp_y_test)
y_preds = predict(clf, Xs)

print_performance(y_trues, y_preds)
```

Next, lets play with each hyperparameter individually to understands its independent effect on the model's performance:

**Penalty:** We notice negligable change in the overfitting of the model. Parameters are already sparse so the effect of L1 regularizaion is very small. 

```python
clf = LinearSVC(dual=False)
penalties = ['l1','l2']

Xs = (yelp_X_bin_train, yelp_X_bin_valid, yelp_X_bin_test)
y_trues = (yelp_y_train, yelp_y_valid, yelp_y_test)

for p in penalties:
    print('\nPenalty: {}'.format(p))
    
    clf.set_params(penalty=p)
    clf.fit(yelp_X_bin_train.tocsc(), yelp_y_train)
    
    y_preds = predict(clf, Xs)

    print_performance(y_trues, y_preds)
```

**Loss Function:** 

```python
clf = LinearSVC()
losses = ['hinge','squared_hinge']

Xs = (yelp_X_bin_train, yelp_X_bin_valid, yelp_X_bin_test)
y_trues = (yelp_y_train, yelp_y_valid, yelp_y_test)

for l in losses:
    print('\nLoss: {}'.format(l))
    
    clf.set_params(loss=l)
    clf.fit(yelp_X_bin_train.tocsc(), yelp_y_train)
    
    y_preds = predict(clf, Xs)

    print_performance(y_trues, y_preds)
```

**Error Term Coefficient:** Lets evaluate the permutation of Error Term Coefficient, Loss and Penalty Function

```python
clf = LinearSVC()

dual = [False, True]
losses = ['squared_hinge','hinge']
penalties = ['l1','l2']
error_coefs = np.linspace(0.0001, 0.005, 15)

Xs = (yelp_X_bin_train, yelp_X_bin_valid)
ys = (yelp_y_train, yelp_y_valid)

for d in dual:
    for l in losses:
        for p in penalties:
            if not d and p == 'l2': continue
            if p == 'l1' and l == 'hinge': continue
            if d and p == 'l1' and l == 'squared_hinge': continue
  
            clf.set_params(dual=d,loss=l,penalty=p)

            train_scores, valid_scores = map_hyperparam(clf, 'C', error_coefs, Xs, ys)

            plot_performance(error_coefs, train_scores, valid_scores, 'Penalty: {} Dual: {} Loss: {} '.format(p,d,l))
```

The most prominent configuration of Hyperparameters is Dual, L2, Square Hinge, C = 0.003
 
Next are its performance metrics

```python
Xs = (yelp_X_bin_train, yelp_X_bin_valid, yelp_X_bin_test)
y_trues = (yelp_y_train, yelp_y_valid, yelp_y_test)

clf = LinearSVC(C=0.003)
clf.fit(yelp_X_bin_train.tocsc(), yelp_y_train)

y_preds = predict(clf, Xs)

print_performance(y_trues, y_preds)
```

**Loss Function:** We notice a slight improvement of performance with squared hinge over hinge loss function

```python
clf = LinearSVC()
losses = ['hinge','squared_hinge']

Xs = (yelp_X_freq_train, yelp_X_freq_valid, yelp_X_freq_test)
y_trues = (yelp_y_train, yelp_y_valid, yelp_y_test)

for l in losses:
    print('\nLoss: {}'.format(l))
    
    clf.set_params(loss=l)
    clf.fit(yelp_X_freq_train.tocsc(), yelp_y_train)
    
    y_preds = predict(clf, Xs)

    print_performance(y_trues, y_preds)
```

**Error Term Coefficient:** Lets evaluate the permutation of Error Term Coefficient, Loss and Penalty Function

```python
clf = LinearSVC()

dual = [False, True]
losses = ['squared_hinge','hinge']
penalties = ['l1','l2']
error_coefs = np.linspace(0.001, 0.01, 15)

Xs = (yelp_X_freq_train, yelp_X_freq_valid)
ys = (yelp_y_train, yelp_y_valid)

for d in dual:
    for l in losses:
        for p in penalties:
            if not d and p == 'l2': continue
            if p == 'l1' and l == 'hinge': continue
            if d and p == 'l1' and l == 'squared_hinge': continue
  
            clf.set_params(dual=d,loss=l,penalty=p)

            train_scores, valid_scores = map_hyperparam(clf, 'C', error_coefs, Xs, ys)

            plot_performance(error_coefs, train_scores, valid_scores, 'Penalty: {} Dual: {} Loss: {} '.format(p,d,l))
```

The 2 most prominent configurations of Hyperparameters are:

 - Dual, hinge loss fn, L2 penalty , C = 0.0072
 - Dual, squared hinge loss fn, L2 penalty , C = 0.0025
 
We choose the Hinge loss function model, as it overfits less the training set and yields better performance over the testing set.

```python
configurations = [{'loss':'hinge', 'C':0.0072},
                  {'loss':'squared_hinge', 'C':0.0025}]

Xs = (yelp_X_freq_train, yelp_X_freq_valid, yelp_X_freq_test)
y_trues = (yelp_y_train, yelp_y_valid, yelp_y_test)

for conf in configurations:
    l = conf['loss']
    c = conf['C']
    
    clf = LinearSVC(loss=l, C=c)

    clf.fit(yelp_X_freq_train.tocsc(), yelp_y_train)

    y_preds = predict(clf, Xs)
    
    print('\nLoss: {}\tC: {}'.format(l,c))
    print_performance(y_trues, y_preds)
```

**Error Term Coefficient:** Lets evaluate the permutation of Error Term Coefficient, Loss and Penalty Function

```python
clf = LinearSVC()

dual = [False, True]
losses = ['squared_hinge','hinge']
penalties = ['l1','l2']
error_coefs = np.linspace(0.001, 0.05, 15)

Xs = (movie_X_freq_train, movie_X_freq_valid)
y_trues = (movie_y_train, movie_y_valid)

for d in dual:
    for l in losses:
        for p in penalties:
            if not d and p == 'l2': continue
            if p == 'l1' and l == 'hinge': continue
            if d and p == 'l1' and l == 'squared_hinge': continue
  
            clf.set_params(dual=d,loss=l,penalty=p)

            train_scores, valid_scores = map_hyperparam(clf, 'C', error_coefs, Xs, ys)

            plot_performance(error_coefs, train_scores, valid_scores, 'Penalty: {} Dual: {} Loss: {} '.format(p,d,l))
```
